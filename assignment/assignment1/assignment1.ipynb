## Task 1: A list of dictionaries

```python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]
```

### Part a

```python
valid_transactions = [
    x for x in transactions if 
        x["price"] != 0 and x["quantity"] >= 0
]

print(valid_transactions)
```

### Part b

```python
for t in valid_transactions:
    t["total"] = t["price"] * t["quantity"]

print(valid_transactions)
```

### Part c

```python
region_rev = {}

for t in valid_transactions:
    region = t["region"]
    revenue = t["total"]

    # Check if there is any region. 
    # The first occurrence of a region therefore results in its revenue.
    # The following occurrences result in the sum of its revenue and the existing amount. 
   
    if region in region_rev:
        region_rev[region] += revenue
    else:
        region_rev[region] = revenue

print(region_rev)
```

### Part d

```python
high_region = max(region_rev, key = region_rev.get)  
high_revenue = region_rev[high_region]

print(f"{high_region} has the highest revenue: {high_revenue}")
```

## Task 2: Creating a function

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]
```

```python
def summarise_sales(N):
    total = sum(N)
    average = round(sum(N)/len(N), 2)
    best_day = N.index(max(N)) + 1    #plus 1 because index starts at 0, day starts at 1
    worst_day = N.index(min(N)) + 1
    above_average_count = sum(1 for i in N if i > average)

    return {
        "total" : total,
        "average" : average,
        "best_day" : best_day,
        "worst_day" : worst_day,
        "above_average_count" : above_average_count
    }

print(summarise_sales(sales))
```

## Task 4: Class and Methods

```python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}

    # Part a
    def add_asset(self, ticker, shares, price_per_share):
        self.holdings[ticker] = {
            "shares": shares,
            "price_per_share": price_per_share
        }

    # Part b
    def total_value(self):
        total = 0
        for data in self.holdings.values():  
            total += data["shares"] * data["price_per_share"]
        return total

    # Part c
    def most_valuable(self):
        return max(
            self.holdings,
            key=lambda t: self.holdings[t]["shares"] * self.holdings[t]["price_per_share"]
        )

    # Part d
    def __str__(self):
        return (f"Portfolio of {self.owner}, "
                f"Assets: {len(self.holdings)}, "
                f"Total value: €{self.total_value()}")

p = Portfolio("Minh Anh")

p.add_asset("AAPL", 10, 150)
p.add_asset("GOOGL", 5, 2800)
p.add_asset("TSLA", 8, 900)

print(p) 

print("Most valuable asset:", p.most_valuable())
```

## Task 8: More Pandas

```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)
```

### Part a

```python
data = {
    "date" : days,
    "visitors" : visitors,
    "conversion_rate" : conversion_rate,
    "avg_order_value" : avg_order_value
}

data = pd.DataFrame(data)
data.set_index("date" , inplace = True)
```

### Part b

```python
data["conversions"] = (data["visitors"] * data["conversion_rate"]).round(0)
data["daily_revenue"] = data["conversions"] * data["avg_order_value"]

print(data)
```
